```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
#Convert into 3D using Principal Component Analysis
iris_t = decomposition.PCA(n_components=3).fit_transform(iris.data)
```

```python
iris_tg = iris.target
```

```python
fig = plt.figure(1)
a = Axes3D(fig)
a.scatter(iris_t[:,0], iris_t[:,1], iris_t[:,2], c = iris_tg)
a.set_title('3D plot for iris')
plt.show()
```
