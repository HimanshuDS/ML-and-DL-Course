```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.optimize import curve_fit
```

```python
#Q1 - Periodic Function for temperature
def yearly_temp(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))
```

```python
tmax = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
tmin = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months = np.arange(12)
fit_max, cv_max = curve_fit(yearly_temp, months, tmax)
fit_min, cv_min = curve_fit(yearly_temp, months, tmin)
```

```python
plt.figure()
plt.plot(months, tmax, 'ro')
plt.plot(months, tmin, 'bo')
plt.plot(months,yearly_temp(months, *fit_max), 'r-')
plt.plot(months, yearly_temp(months, *fit_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature (oC)')
plt.show()
```

```python
#Q2 Matplotlib visualization
url = 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
pd.Categorical(titanic['sex'])
```

```python
m = np.asarray(np.where(titanic['sex'] == 'male'))
```

```python
f = np.asarray(np.where(titanic['sex'] == 'female'))
```

```python
cat = ['Male', 'Female']
val = [m.size, f.size]
fig, ax = plt.subplots()
ax.pie(val, labels = cat, autopct='%1.1f%%')
plt.show()
```
